```python
# Imports
from music21 import *
from IPython.display import display
import zipfile
import pickle
import numpy as np
```

```python
# Absolute path to a file -- this is a local file and is not on github
path_to_all_vg_midis = r'C:\Dev\python\CS4850\TEAM2_LOCAL_DATA\All_Castlevania'
midi_fnames = os.listdir(path_to_all_vg_midis)
```

```python
%%time

# Transpose to key of C
majors = dict([
    ('A-', 4), ('G#', 4),('A', 3),
    ('A#', 2),('B-', 2),('B', 1),
    ('C', 0),('C#', -1),('D-', -1),
    ('D', -2),('D#', -3),('E-', -3),
    ('E', -4), ('F', -5),('F#', 6),
    ('G-', 6), ('G', 5)
])
minors = dict([(
    'G#', 1), ('A-', 1),('A', 0),
    ('A#', -1),('B-', -1),('B', -2),
    ('C', -3),('C#', -4),('D-', -4),
    ('D', -5),('D#', 6),('E-', 6),
    ('E', 5),('F', 4),('F#', 3),
    ('G-', 3),('G', 2)
])

# Tranpose the scores
transposed_to_key_of_c_piano_stream_list = []
for file_ in midi_fnames:

    # Log file
    print('Processing', file_)

    # Parse the file
    score = converter.parse(os.path.join(path_to_all_vg_midis, file_))

    # Instrument_score
    instrument_score = instrument.partitionByInstrument(score)

    if ('Piano' in [part.partName for part in instrument_score.parts]):

        # instrument score is now piano
        instrument_score = instrument_score['Piano']

        # Find the key of the score
        key = score.analyze('key')

        # Log current key and tonic
        print ('Key info of score', key.tonic.name, key.mode)


        try:

            # Find key of instrument stream
            key = instrument_score.analyze('key')

            # Log key and tonic
            print ('Key info of instrument_score', key.tonic.name, key.mode)

            # Check mode
            if key.mode == "major":
                halfSteps = majors[key.tonic.name]
                
            elif key.mode == "minor":
                halfSteps = minors[key.tonic.name]
            
            # Transpose the new score
            newscore = instrument_score.transpose(halfSteps).chordify()

            # Reset the key
            key = newscore.analyze('key')

            print(type(key.name))
            print(key.name)

            # Log key
            print ('Key info after transposition of instrument_score:', key.tonic.name, key.mode)

            # Append score to list
            valid_key_names = ['c major', 'C major', 'A minor', 'a minor']
            if (key.name in valid_key_names):
                print('Transposed correctly. Appending to list.')
                transposed_to_key_of_c_piano_stream_list.append(newscore)

            else:
                print('Transposed incorrectly.')
        
        except exceptions21.Music21Exception:
            print('Music21 Exception')

    else:

        # Log
        print('No Piano')
        
    # Whitespace
    print()
```

```python
# Inspect scores
print('Number of piano songs successfully transposed to key of C:', len(transposed_to_key_of_c_piano_stream_list))
```

```python
%%time

# Write streams as tests
for ix, stream in enumerate(transposed_to_key_of_c_piano_stream_list):
    print('Writing to file.')
    stream.write('midi', os.path.join(r'C:\Dev\python\CS4850\TEAM2_LOCAL_DATA\Transposed_Songs', '_' + str(ix) + '.mid'))
```

```python
# Pickle the scores obj
with open('../pickled_data/pickled_tentatively_transposed_instrument_scores', 'wb') as fobj:
    pickle.dump(transposed_to_key_of_c_piano_stream_list, fobj)
```
